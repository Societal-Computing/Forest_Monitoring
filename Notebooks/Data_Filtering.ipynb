##counting polygons Per site,Estimating polgon areas

```python
def count_all_polygons(all_geom_data):
    try:
        all_geom_data = ast.literal_eval(all_geom_data)
        count_polygons = 0
        polygon_areas = []
        polygon_geometries = []

        project = pyproj.Transformer.from_crs(
            pyproj.CRS('EPSG:4326'),
            pyproj.CRS('EPSG:6933'),
            always_xy=True).transform

        for geom_data in all_geom_data:
            if geom_data["geometry"]["type"] == "Polygon":
                count_polygons += 1
                polygon = shape(geom_data["geometry"])
                projected_polygon = transform(project, polygon)
                polygon_areas.append(projected_polygon.area)
                polygon_geometries.append(polygon)
            if geom_data["geometry"]["type"] == "MultiPolygon":
                count_polygons += len(geom_data["geometry"]["coordinates"])
                for polygon_coordinates in geom_data["geometry"]["coordinates"]:
                    polygon = shape({"type": "Polygon", "coordinates": polygon_coordinates})
                    projected_polygon = transform(project, polygon)
                    polygon_areas.append(projected_polygon.area)
                    polygon_geometries.append(polygon)

        return count_polygons, polygon_areas, polygon_geometries
    except Exception as e:
        print(f"Error: {e}")
        return None, None, None

df_with_site_polygon['polygon_count'], df_with_site_polygon['polygon_areas'], df_with_site_polygon['polygons'] = zip(*df_with_site_polygon.apply(lambda row: count_all_polygons(row['sites']), axis=1))
```

### Removing Outliers

```python
# Calculate Q1, Q2 and IQR
Q1 = df_sorted['trees_per_acre'].quantile(0.25)
Q3 = df_sorted['trees_per_acre'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df_sorted[(df_sorted['trees_per_acre'] < lower_bound) | (df_sorted['trees_per_acre'] > upper_bound)]

# Drop outliers
df_sorted_outliers_drp = df_sorted[(df_sorted['trees_per_acre'] >= lower_bound) & (df_sorted['trees_per_acre'] <= upper_bound)]
```

# Downloading the 2d Images From Plant for planet

```python
import os

# Check if the 'images' directory exists
if not os.path.exists('images'):
    # If not, create the directory
    os.makedirs('images')
```

```python
import requests
import json
import os
import shutil

# URL of the website
url = "https://app.plant-for-the-planet.org/app/projects?_scope=map&currency=EUR&tenant=ten_NxJq55pm&filter[purpose]=trees,conservation&locale=en"

# Base URL of the website
# base_url = "https://app.plant-for-the-planet.org/"
base_url = "https://cdn.plant-for-the-planet.org/media/cache/project/medium/"

# Send a GET request to the website
response = requests.get(url)

# Parse the JSON content
projects = json.loads(response.text)

# Loop through each project
for project in projects:
    # Extract the image URL
    image_url = base_url + project['properties']['image']

    # Send a GET request to the image URL
    response = requests.get(image_url, stream=True)



    # Check if the request is successful
    if response.status_code == 200:
        # Open a local file with write-binary mode
        with open(os.path.join('images', project['properties']['id'] + '.jpg'), 'wb') as out_file:
            # Write the content of the image to the file
            shutil.copyfileobj(response.raw, out_file)

# Delete the image memory
del response
```

```python
import requests
import json
import os
import shutil

# URL of the website
url = "https://app.plant-for-the-planet.org/app/projects?_scope=map&currency=EUR&tenant=ten_NxJq55pm&filter[purpose]=trees,conservation&locale=en"

# Base URL of the website
# base_url = "https://app.plant-for-the-planet.org/"
base_url = "https://cdn.plant-for-the-planet.org/media/cache/project/medium/"

# Send a GET request to the website
response = requests.get(url)

# Parse the JSON content
projects = json.loads(response.text)

# Loop through each project
for project in projects:
    # Extract the image URL
    image_url = base_url + project['properties']['image']

    # Send a GET request to the image URL
    response = requests.get(image_url, stream=True)



    # Check if the request is successful
    if response.status_code == 200:
        # Open a local file with write-binary mode
        with open(os.path.join('images', project['properties']['id'] + '.png'), 'wb') as out_file:
            # Write the content of the image to the file
            shutil.copyfileobj(response.raw, out_file)

# Delete the image memory
del response
```
