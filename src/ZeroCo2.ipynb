Extract data  data from https://zeroco2.eco/en/projects/

```python
import json
import pandas as pd
with open('../input/ZeroCo2/ZeroCo2.json', 'r') as file:
    data = json.load(file)

print(type(data))


print(data[0])


if isinstance(data, list) and isinstance(data[0], dict):
    zc_df = pd.DataFrame(data)
    zc_df.info()
```

```python
zc_df.head()
```

```python
# Renaming the columns to follow our naming format in the paper columns section
columns_rename_mapping = {
    'project_id': 'project_id_reported',
     'id': 'site_id_reported',
    'gps_position.coordinates': 'geometry_reported',
    'description': 'description_reported',
    'countryName': 'country',
    'date_created' : 'planting_date_reported',
    #'data.forests.main_image': 'url',
    #'survivalRate': 'survival_rate_reported',
}

df = zc_df[list(columns_rename_mapping.keys())]


df.rename(columns=columns_rename_mapping, inplace=True)
```

```python
#df['site_id_reported'] = ['tist_asia_{}'.format(i) for i in range(1, len(asia_df) + 1)]
df = df.assign(site_sqkm=None,species_count_reported=None, species_planted_reported=None, survival_rate_reported=None,trees_planted_reported=None,country=None)
df.info()
```

```python
df.to_csv('../input/ZeroCo2/ZeroCo2.csv', index=False)
```
