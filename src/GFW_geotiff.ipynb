```python
import requests
import geopandas as gpd
import rasterio
from shapely.geometry import box, shape
import os
import json

tile_id_url = "https://services2.arcgis.com/g8WusZB13b9OegfU/arcgis/rest/services/Tropical_Tree_Cover/FeatureServer/0/query?outFields=*&where=1%3D1&f=geojson"


response = requests.get(tile_id_url)


if response.status_code == 200:
  
    data = json.loads(response.text)

    # Extracting tropical tree cover the tile IDs
    tile_ids = [feature['properties']['tile_id'] for feature in data['features']]
else:
    print(f"Request for tile IDs failed with status code {response.status_code}")
    tile_ids = []

url = "https://data-api.globalforestwatch.org/dataset/wri_tropical_tree_cover/v2020/download/geotiff"

headers = {
    'x-api-key': 'da26c443-4f75-4a92-a437-611ff7c4c13a'
}

# Directory containing the GeoJSON files
directory = "/home/idisc02/Saarland_Forest_monitoring_research/Forest_Monitoring/input/Tree_Nation_Polygons/tree_nation_small_size_polygons/Clear_Change_Polygons/"

# Directory to save the output files
output_directory = "/home/idisc02/Saarland_Forest_monitoring_research/Forest_Monitoring/input/Tree_Nation_Polygons/tree_nation_small_size_polygons/Clear_Change_Polygons/"

# Looping over each file in the directory
for filename in os.listdir(directory):
    if filename.endswith(".geojson"):
        # Loading the GeoJSON file
        geojson = gpd.read_file(os.path.join(directory, filename))

        for tile_id in tile_ids:
            params = {
                "grid": "10/40000",  
                "tile_id": tile_id,  
                "pixel_meaning": "decile"  
            }

      
            response = requests.get(url, params=params, headers=headers)

            
            if response.status_code == 200:
         
                with open('temp.tif', 'wb') as f:
                    f.write(response.content)
                
                # Loading the temporary GeoTIFF
                with rasterio.open('temp.tif') as src:
                    geotiff = src.read()

                # Checking if the GeoTIFF intersects with the GeoJSON
                if any(shape(geojson.geometry).intersects(box(*src.bounds)) for geojson in geojson.itertuples()):
                    # If they intersect, save the file with the tile ID as the name
                    with open(os.path.join(output_directory, f'{tile_id}_output_file.tif'), 'wb') as f:
                        f.write(response.content)
            else:
                print(f"Request for tile {tile_id} failed with status code {response.status_code}")
```
