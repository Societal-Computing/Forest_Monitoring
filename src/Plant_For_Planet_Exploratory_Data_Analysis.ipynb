Function to Count sites for each project with complete polygons information

```python
import ast
def count_all_poylgons(all_geom_data):
  try:
    all_geom_data = ast.literal_eval(all_geom_data)
    print(type(all_geom_data))
    count_polygons = 0

    for geom_data in all_geom_data:
      #geom_data = dict(geom_data)
      if geom_data["geometry"]["type"]=="Polygon":
          count_polygons += 1
      if geom_data["geometry"]["type"]=="MultiPolygon":
          count_polygons = count_polygons + len(geom_data["geometry"]["coordinates"])
    return count_polygons
  except Exception as e:
    print(all_geom_data)
    return  None
```

```python
df_with_site_polygon['polygon_count'] = df_with_site_polygon.apply(lambda row: count_all_poylgons(row['sites']), axis=1)
```

```python
import requests
import json
import os
import shutil

# URL of the website
url = "https://app.plant-for-the-planet.org/app/projects?_scope=map&currency=EUR&tenant=ten_NxJq55pm&filter[purpose]=trees,conservation&locale=en"

# Base URL of the website
# base_url = "https://app.plant-for-the-planet.org/"
base_url = "https://cdn.plant-for-the-planet.org/media/cache/project/medium/"

# Send a GET request to the website
response = requests.get(url)

# Parse the JSON content
projects = json.loads(response.text)

# Loop through each project
for project in projects:
    # Extract the image URL
    image_url = base_url + project['properties']['image']

    # Send a GET request to the image URL
    response = requests.get(image_url, stream=True)



    # Check if the request is successful
    if response.status_code == 200:
        # Open a local file with write-binary mode
        with open(os.path.join('images', project['properties']['id'] + '.jpg'), 'wb') as out_file:
            # Write the content of the image to the file
            shutil.copyfileobj(response.raw, out_file)

# Delete the image memory
del response
```
